## Abstract


<p><div style="text-align: justify"> Covid-19 crisis has been changing many aspects of individuals' social life globally with the limitation of group gatherings, economic downturns, and the nervousness caused by the fear of getting the virus. We focus our study on a specific group of people - univerisity students - in order to investigate the impact of Covid-19 crisis on the self-assessed studying motivation, group project performance, exam performance, feeling of loneliess, and financial stress, by conducting an online survey. Our results show that the pandemic significantly influences students' studying motivation, feeling of loneliness, and financial stress. Furthermore, we concentrate this research on a subgroup - students in Dutch universities - and we found evidence that the Covid-19 crisis significantly decreases the self-assessed studying movation, group project performance, and increases the feeling of loneliness of the students in Dutch universities. </div></p>

## Research questions

In this assignment our aim is to answer the following 5 research questions related to students' life during the Covid-19 crisis:
1. Does the Covid-19 crisis affect the university students' self-assessed **studying motivation** ? 
2. Does the Covid-19 crisis affect the university students' self-assessed **groupwork performance**?
3. Does the Covid-19 crisis affect the university students' self-assessed **exam performance**?
4. Does the Covid-19 crisis affect the univeristy students' self-assessed **loneliness** ? 
5. Does the Covid-19 crisis affect the university students' self-assessed **financial stress**? 



## Motivation
<p><div style="text-align: justify">
Covid-19 crisis has impacted many aspects of the societies, and some might be even profound. Dealing with univerity students, the lockdown prevents face-to-face teaching lessons, and exams are mostly online. In addition, the limitation about the maximum number of people in a group (maximum 2 people in the Netherlands) impedes face-to-face debate, which might reduces the efficiency of group projects' discussions. The lockdown measurements could have psychological impact on students as well. Lastly, many countries are going through economic downturn during this period. For example, the 27 countries of the European Union (EU) suffered from -11.4 % recession in the second quarter of 2020, that is when the Covid-19 was most severe (Eurostat, 2020). The economic downturn could potentially influence the financial stress of univerity students, because they might lose their source of income (i.e. from part-time jobs). </div></p>

<p><div style="text-align: justify"> Some studies have already investigated the effects of Covid-19 on university students. In the US, Aucejo et al. (2020) have conducted a survey among 1,500 undergraduate students at Arizona State University in April 2020. They found that 13% of students have delayed graduation, 40% lost a job or an internship, and 29% expect to earn less at the age of 35. Moreover, they pointed out that the effects of Covid-19 pandemic have been highly heterogeneous among university students: in fact, the effect on delayed graduation is 55% more pronounced for lower-income students than higher-income ones. In addition, according to the SMaRteN study (2020), 66% of the 4,000 doctoral candidates who answered the survey, are worried about their future, while 70% of the respondents are facing financial stress. Furthermore, Ambreen and Utsav collected the responses of 500 Indian students, and they found that the 70% is stressed about their studies, and almost 80% of students are not satisfied with online education. </div></p>

<p><div style="text-align: justify"> Following this path, we are interested to further investigate the impact of the pandemic on students' self-assessed studying motivation, performance, feeling of loneliness, and financial stress. These factors could potentially impact the long-term development of university students, such as applying post-graduate schools, future earnings, and satisfaction of their future careers. </div></p>

#### Data analysis (Methodology)
<p><div style="text-align: justify"> 
We apply Ordinary Least Squares (OLS) regression models to study the causal impact of Covid-19 on self-assessed outcome variables. Two types of OLS regression models are employed. The models without control variables are written as: </div></p>

<center>$Y_{i}=\beta_0+\beta_1 Covid_{i}+\epsilon_{i}$
</center>

<p><div style="text-align: justify"> The models with individual background control variables are written as: </div></p>

<center>$Y_{i}=\beta_0+\beta_1 Covid_{i}+ \beta_2 X_{i} + \epsilon_{i}$
</center>

<p><div style="text-align: justify"> 
In both equations above, $Y_{i}$ is the self-assessed level of motivation, group project performance, exam performance, loneliness, and financial stress before or during the Covid-19 crisis. $Covid_{i}$ represents whether the observation refers to the pre-Covid period or during the pandemic. The effect of Covid is shown by the coefficient $\beta_1$. In the second equation, $X_{i}$ is a vector of individual background control variables, including gender, nationality, domicile, relationship status, household composition, level of education, major of studying, and the country where the respondent's university is located. $\epsilon_{i}$ is the error term.
We compare the results of two types of models for robustness test. </div></p>

## Main assumptions
<p><div style="text-align: justify"> In order to assess the causal impact of Covid-19 on students' outcomes, we have to know the students' outcomes with and without the Covid-19 crisis, mathematically, the potential outcome framework is:</div></p>

<center> $\Delta(Y)_{i}=Y_{i}(Covid19=1)-Y_{i}(Covid19=0)$ </center>

<p><div style="text-align: justify"> The first term on the right hand side stands for the student i' outcomes with Covid-19.The second term on the right hand side is the student i' outcomes in the state without Covid-19, which is unoberserved. </div></p>

<p><div style="text-align: justify"> This study is based on two main assumptions: </div></p>

<p><div style="text-align: justify"> 1) We use students' self-reported outcomes before the Covid-19 crisis as counterfactual to represent students' outcomes without Covid-19. </div></p>

<p><div style="text-align: justify"> 2) Our results are fully based on the self-assessed survey. Therefore, another assumption is that students report their outcomes and background information correctly. Especially, we expect they have good understanding of our scales of outcome variables. </div></p>

<p><div style="text-align: justify">
The limitation of these assumption are discussed in the last section. </div></p>

#### More data cleaning

<p><div style="text-align: justify"> Furthermore, before running the regressions, we need to merge together the *pre*-Covid columns with the *post*-Covid ones. For example, we would like to have a dataset in which we have just one column called "Motivation", instead of "Motivation_before" and "Motivation_after". The same hold for: Group, Exam, Loneliness, Financial_stress, and Income. However, we need to create a new column called "Covid" in order to know which observations belong to the *pre*-Covid period and which ones to the *post*-Covid. This new variable should assume the value of "0" for all the *pre*-Covid observations, and "1" for all the *post*-Covid.</div></p>

<p><div style="text-align: justify"> In the following chunks, we explain step by step how we did this. First, we created 2 copies of the original dataframe, called respectively: "df_covid_0" and "df_covid_1". However, these two are not exact copies. In fact, both have the socio-demographic characteristics, but df_covid_0 also has the *before* variables, while df_covid_1 has the *after* variables. </div></p>

```python
# Create a copy of df and name it df_covid_0 (socio-demographic + before-Covid variables)
df_covid_0 = df[["Motivation_before", "Group_before", "Exam_before", "Loneliness_before", "Financial_stress_before", 
                 "Income_before", "Student_finance", "Gender", "Nationality", "Second_nationality",
                 "Domicile", "Relationship_status", "Household_composition", "Education", "Major", "Second_major",
                 "University_country"]].copy()

# Create a copy of df and name it df_covid_1 (socio-demographic + after-Covid variables)
df_covid_1 = df[["Motivation_after", "Group_after", "Exam_after", "Loneliness_after", "Financial_stress_after",
             "Income_after", "Student_finance", "Gender", "Nationality", "Second_nationality",
             "Domicile", "Relationship_status", "Household_composition", "Education", "Major", "Second_major",
             "University_country"]].copy()
```

<p><div style="text-align: justify"> Second, we create a new column called "Covid" in both the dataframes. This variable has the value of "0" for all the observations that are in df_covid_0, and the value of "1" for all the observations that are present in df_covid_1. </div></p>

```python
# Create the column "Covid" which is 0 for all the observations
df_covid_0['Covid'] = 0

# Create the column "Covid" which is 1 for all the observations
df_covid_1['Covid'] = 1
```

<p><div style="text-align: justify"> Third, we rename the columns of both the dataframes in order to give them the same names. We want to merge them in the end, therefore this step is necessary. </div></p>

```python
# Provide the columns of both the df with the same names
df_covid_0.columns = ["Motivation", "Group", "Exam", "Loneliness", "Financial_stress",
                      "Income", "Student_finance", "Gender", "Nationality", "Second_nationality",
                      "Domicile", "Relationship_status", "Household_composition", "Education", "Major", "Second_major",
                      "University_country", "Covid"]

df_covid_1.columns = ["Motivation", "Group", "Exam", "Loneliness", "Financial_stress",
                      "Income", "Student_finance", "Gender", "Nationality", "Second_nationality",
                      "Domicile", "Relationship_status", "Household_composition", "Education", "Major", "Second_major",
                      "University_country", "Covid"]
```

<p><div style="text-align: justify"> At this point, we have 2 dataframes, df_covid_0 and df_covid_1. These two have the same variables, but the first one has all the *pre*-Covid observations, while the second one has all the *post*_Covid ones. And both have the same socio-demographic characteristics. Therefore, we can merge them now. </div></p>

```python
# Merge df_covid_0 and df_covid_1
df_total = df_covid_0.append(df_covid_1)
```

## Limitation and Discussion
<p><div style="text-align: justify"> 
The significant impact of Covid-19 on many aspects of students' life was expected, and it is difficult to avoid. Many univerisities have already made their best to continue with an effective and efficient teaching process. Regarding the loneliness and financial stress of the students, our results encourage universities and governments to provide more support for these young people. For example, universities could employ more psychologists to shorten the period of waiting for students with mental difficulties. Governments could provide more generous students finance or loans during the Covid-19 lockdown periods (although in the Netherlands, we found no significant effect of Covid-19 on financial stress of the students, so this might not applied to the Dutch government).</div></p>

<p><div style="text-align: justify"> 
In the end, we stress some major limitations of our study. Due to the time limitation, we were unable to collect larger sample for higher level of accurary of the estimation. Besides, it should be noticed that our results are based on the assumption that students assessed their studying motivation, group project performance, exam performance, feeling of loneliness, and financial stress correctly. In particular, they should provide the above information about the whole period of the Covid-19 crisis during a certain point of time (the time when they fill in the survey). For example, if a student just failed one exam, he might underestimate his or her overall exam performance. The background information is also provided by students, thus we rely on the correctness of this infomation as well.</div></p>

<p><div style="text-align: justify"> 
Furthermore, we use the self-assessed outcomes of the students before the Covid-19 crisis as a counterfactual to represent the expected outcomes in the state without Covid-19 crisis. However, many other external factors change during the Covid-19 crisis, compared to the time before the Covid-19 crisis, such as the university courses, and other government policies.We could not take them into account.</div></p>

<p><div style="text-align: justify"> 
Future studies can go a step further, and study the impact of Covid-19 on long-term development of university students, such as the time intervel between graduation and finding the first job, and the job statisfaction of the students.</div></p>
